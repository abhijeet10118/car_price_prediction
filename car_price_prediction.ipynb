```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) 
```

```python
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit([['name','company','fuel_type']])
# ohe.categories_
```

```python
tranformed_columns = make_column_transformer(
    (ohe, ['name', 'company', 'fuel_type']),
    remainder='passthrough'
)
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(tranformed_columns,lr)
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(tranformed_columns,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
    prediction_r2_score=r2_score(y_test,y_pred)
    print(f"iteration {i} complete , the r2score is {prediction_r2_score}")
```

```python
import numpy as np
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(tranformed_columns,lr)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
